```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# lamda = 1
r = 3.8
lds = np.linspace(0,5,10000)
N = 500
x = .6 + np.zeros(N)
endcap = np.arange(round(N*.9),N)
for ldi in range(len(lds)):
  for n in range(N-1):
    if 0 <= x[n] < 0.5:
      x[n+1] = (2*r*lds[ldi]*x[n])/(1+lds[ldi]*(1-2*x[n])**2) - math.floor((2*r*lds[ldi]*x[n])/(1+lds[ldi]*(1-2*x[n])**2))
    elif 0.5 <= x[n] < 1:
      x[n+1] = (r*lds[ldi]*(2*x[n]-1))/(1+lds[ldi]*(2-2*x[n])**2) - math.floor((r*lds[ldi]*(2*x[n]-1))/(1+lds[ldi]*(2-2*x[n])**2))
  u = np.unique(x[endcap])
  ld = lds[ldi]*np.ones(len(u))
  plt.plot(ld,u,'k.', markersize=1)
# plt.plot(x,'s-')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

R = np.linspace(1,4,10000)
# lamda = 1
r = 3.8
N = 500
LE = []
result = []
for lamda in R:
    x = 0.6
    for n in range(N):
      if 0 <= x < 0.5:
        x = (2*lamda*r*x)/(1+lamda*(1-2*x)**2) - math.floor((2*lamda*r*x)/(1+lamda*(1-2*x)**2))
      elif 0.5 <= x < 1:
        x = (lamda*r*(2*x-1))/(1+lamda*(2-2*x)**2) - math.floor((lamda*r*(2*x-1))/(1+lamda*(2-2*x)**2))
    for n in range(N):
      if 0 <= x < 0.5:
        x = (2*lamda*r*x)/(1+lamda*(1-2*x)**2) - math.floor((2*lamda*r*x)/(1+lamda*(1-2*x)**2))
        result.append(np.log(abs((2*lamda*r*(1+lamda*(1-2*x)**2)-((-4*lamda+8*lamda*x)*(r*lamda*2*x)))/((1+lamda*(1-2*x)**2)**2))))
      elif 0.5 <= x < 1:
        x = (lamda*r*(2*x-1))/(1+lamda*(2-2*x)**2) - math.floor((lamda*r*(2*x-1))/(1+lamda*(2-2*x)**2))
        result.append(np.log(abs((2*lamda*r*(1+lamda*(2-2*x)**2)-((-8*lamda+8*lamda*x)*(r*lamda*(2*x-1))))/((1+lamda*(1-2*x)**2)**2))))
    LE.append(np.mean(result))
plt.figure(figsize=(10,10))
plt.grid('on')
plt.plot(R, LE, ls='', marker='.', alpha=1)
plt.show()
```
